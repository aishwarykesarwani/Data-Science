```python
##- sepal length in cm
   # - sepal width in cm
  #  - petal length in cm
 #   - petal width in cm
```

```python
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width','petal length','petal width'])
```

```python
y=iris['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()
```

```python

treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python

y_pred=treeclassifier.predict(X_test)
```

```python

from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

### HyperParameter Tuning

```python
## doing post pruning
## good for small dataset

mode_post=DecisionTreeClassifier(max_depth=2)
```

```python
mode_post.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(mode_post,filled=True)
```

```python
## to avoid overfitting
## pre pruning

param={
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

treemodel=DecisionTreeClassifier()
```

```python
grid=GridSearchCV(treeclassifier,param_grid=param,cv=5,scoring='accuracy')
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python

grid.best_params_
```

```python
grid.best_score_

y_pred=grid.predict(X_test)
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
