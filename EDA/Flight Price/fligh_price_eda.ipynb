### Making a custom function to convert (xh ym) to minutes (also handling cases like(xh) or (xm))

```python
def func(df):
    try:# cases like 19h or 10m
        a,b=df.split(' ');
    except:
        if(df[-1]=='h'):
            return int(df[0:len(df)-1])*60
        else:
            return int(df[0:len(df)-1])
    a=a[:len(a)-1]
    b=b[:len(b)-1]
    if(len(b)==0):
        b=0;
    else :
        b=int(b);
    if(len(a)==0):
        a=0;
    else :
        a=int(a);
    # print(a,b)
    # print(type(a),type(b))
    return a*60+b
```
